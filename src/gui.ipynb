```python
def plotAction(wdgt):
  with plot_output: 
    clear_output()
    waterway_type1 = waterway_type.value[0]
    waterway_type2 = waterway_type.value[1]
    userPath = user_path.value
    filenm = select_filenm.value.split('.')[0]
    city_name = filenm.split('_')[0].title()
    userPlotBy = user_plotby.value
    plotvisual(city_name, waterway_type1, waterway_type2, userPath, filenm, userPlotBy)


def barplotAction(wdgt):
  with barplot_output: 
    clear_output()
   
    userPath = user_path.value
    filenm = select_filenm.value.split('.')[0]
    city_name = filenm.split('_')[0].title()

    plt = barplotvisual(city_name, userPath, filenm)
    plt.show()


def saveJsonAction(wdgt):
  if len(select_manualLocation.value) == 0: 
    print(f'Querying Results for {select_prefecture.value}')
    city_name = (', ').join([select_prefecture.value + " Prefecture", 'Japan'])
  else: 
    print(f'Querying Results for {select_manualLocation.value}')
    city_name = select_manualLocation.value
  waterway_type1 = waterway_type.value[0]
  waterway_type2 = waterway_type.value[1]
  userPath = user_path.value
  filenm = user_filenm.value
  saveJSON(city_name, waterway_type1, waterway_type2, userPath, filenm)

def geoDataAction(wdgt):
  
  input_path = user_path.value
  filenm = select_filenm.value.split('.')[0]
  dataType = user_geodata.value
  output_path = user_geoDataPath.value

  savegeoData_as(input_path, filenm, dataType, output_path)


def saveBarPlotAction(wdgt):
  input_path = user_path.value
  filenm = select_filenm.value.split('.')[0]
  city_name = filenm.split('_')[0].title()
  output_path = user_barPlotPath.value

  plt = barplotvisual(city_name, input_path, filenm)
  plt.savefig(f'{output_path}/{filenm}.png')

def refreshListAction(wdgt):
  select_filenm.options = getFiles_asList(user_path)
```

```python
header = widgets.HTML("<h1>OSM Feature Search By Tag</h1>", layout=Layout(height='auto'))
header.style.text_align='center'

select_prefecture = widgets.Dropdown(
    options=list(jp_prefectures.values()),
    value='None',
    description='Select Prefecture',
    disabled=False, style=dict(description_width='initial')
)

select_manualLocation = widgets.Text(
    value='',
    placeholder='enter city',
    description='OR Enter Location:',
    disabled=False, style=dict(description_width='initial')
    )

waterway_type = widgets.SelectMultiple(
    options=['stream', 'river', 'canal', 'tidal_channel', 'ditch', 'drain', 'riverbank'],
    value=['stream', 'river'],
    description='waterway_type',
    disabled=False, style=dict(description_width='initial')
)

user_path = widgets.Text(
    value='data/json',
    placeholder='enter path',
    description='Save Location',
    disabled=False, style=dict(description_width='initial')
    )

user_filenm = widgets.Text(
    value= '',
    placeholder='enter filename',
    description='Save file as',
    disabled=False, style=dict(description_width='initial')
    )

user_plotby = widgets.RadioButtons(
    options=['layer', 'source'],
    description='Plot By:',
    disabled=False
)


select_filenm = widgets.Dropdown(
    options=getFiles_asList(user_path),
    description='Select file',
    disabled=False, style=dict(description_width='initial')
)


user_geodata = widgets.ToggleButtons(
    options=['geojson', 'shapefile'],
    description='Save as:',
    disabled=False
)

user_geoDataPath = widgets.Text(
    value='data/geo_data',
    placeholder='enter path',
    description='Save File Path',
    disabled=False, style=dict(description_width='initial')
    )

user_barPlotPath = widgets.Text(
    value='data/plots',
    placeholder='enter path',
    description='Save File Path',
    disabled=False, style=dict(description_width='initial')
    )


plot_output = widgets.Output(layout = Layout(width='600px', height='auto'))

barplot_output = widgets.Output(layout = Layout(width='600px', height='auto'))
```

```python
"""#### Buttons"""

# Save as shp or geojson 
Save_geoData_button = widgets.Button(
    description='Download',
    disabled=False,
    button_style='success',
    layout = Layout(width='150px', height='auto'),
    style=dict(description_width='initial'))

Save_geoData_button.on_click(geoDataAction)


# Save plot as a png 
Save_barplot_button = widgets.Button(
    description='Download',
    disabled=False,
    button_style='success',
    layout = Layout(width='100px', height='auto'),
    style=dict(description_width='initial'))

Save_barplot_button.on_click(saveBarPlotAction)

# Show Map 
Plot_button = widgets.Button(
    description='Map Data',
    disabled=False,
    button_style='warning',
    icon='signal', layout = Layout(width='100px', height='25px'),
    style=dict(description_width='initial'))

Plot_button.on_click(plotAction)

# Save JSON file 
Save_JSON_button = widgets.Button(
    description='Query & Save JSON',
    disabled=False,
    button_style='info',
    icon='signal', layout = Layout(width='200px', height='auto'),
    style=dict(description_width='initial'))

Save_JSON_button.on_click(saveJsonAction)


# Show Bar Plot 
barPlot_button = widgets.Button(
    description='Plot Data',
    disabled=False,
    button_style='warning',
    icon='danger', layout = Layout(width='100px', height='auto'),
    style=dict(description_width='initial'))

barPlot_button.on_click(barplotAction)


# Show Bar Plot 
refreshFileList_button = widgets.Button(
    description='Refresh',
    disabled=False,
    button_style='info',
    icon='danger', layout = Layout(width='100px', height='auto'),
    style=dict(description_width='initial'))

refreshFileList_button.on_click(refreshListAction)
```

```python
tab1 = HBox(children=[VBox(children=[tab1_header, #tab1_description, tab1_directions, tab1_directions_text, 
                                     HBox(children=[select_prefecture, select_manualLocation]), 
                                     waterway_type, 
                                     HBox(children=[user_path, user_filenm]),
                                     HBox(children=[Save_JSON_button])
                                     ])])
tab2 = VBox(children=[HBox(children=[select_filenm, refreshFileList_button]), user_plotby , plot_output, Plot_button, HBox(children=[user_geoDataPath, user_geodata ]), HBox(children=[Save_geoData_button])])

tab3 = VBox(children=[select_filenm, barplot_output, user_barPlotPath, HBox(children=[barPlot_button, Save_barplot_button])])

tab_nest = widgets.Tab()
tab_nest.children = [tab1, tab2, tab3]
tab_nest.set_title(0, 'Query OSM Data')
tab_nest.set_title(1, 'Map Visualization')
tab_nest.set_title(2, 'Plot Layer Count')
```
